## Data Wrangling

1. Read the mouse RNA-seq gene_expression data and mouse Neuroelectro ephys data into dataframes

```python
## YOUR CODE HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION
mouse_rnaseq_df = pd.read_csv('data/mouse_rna_seq/mouse_gene_expression.xls')     # Read mouse RNA-seq data into a DataFrame
mouse_ephys_df = pd.read_excel('data/neuroelectro/neurophysiology_data.xlsx')     # Read mouse Neuroelectro ephys data into a DataFrame
mouse_rnaseq_df = mouse_rnaseq_df.set_index('feature').T     # Set the index for the RNA-seq DataFrame as the features and take the transpose
```

```python
mouse_rnaseq_df.head()     # View a few rows of mouse_rnaseq_df
```

```python
mouse_rnaseq_df = mouse_rnaseq_df.loc[:, mouse_rnaseq_df.sum() != 0]    # Removing columns with all entries 0
normalized_rnaseq_df = (mouse_rnaseq_df - mouse_rnaseq_df.mean()) / mouse_rnaseq_df.std()     # Normalization of RNA-seq data
normalized_rnaseq_df.head()
```

```python
rnaseq_cell_list = list(normalized_rnaseq_df.index)
CA1_samples = []
ENT_samples = []
for cell in rnaseq_cell_list:
    if ('CA1' in cell) and ('-' not in cell):
        CA1_samples.append(cell)
    elif ('ENTm' in cell) and ('L3' in cell):
        ENT_samples.append(cell)
CA1_rnaseq_data = normalized_rnaseq_df.loc[CA1_samples]     # Only keeping RNA-seq data for CA1
ENT_rnaseq_data = normalized_rnaseq_df.loc[ENT_samples]     # Only keeping RNA-seq data for Medial Entorhinal Cortex
```

```python
CA1_rnaseq_data.head()
```

```python
ENT_rnaseq_data.head()
```

```python
mouse_ephys_df.head()   # View a few rows of mouse_ephys_df
```

```python
CA1_ephys_df = mouse_ephys_df[mouse_ephys_df['NeuronType'] == 'Hippocampus CA1 pyramidal cell']     # Only keeping data relevant to CA1
CA1_ephys_df = (CA1_ephys_df.dropna(how='all', axis=1).select_dtypes(exclude=['object'])     # Drop columns with strings, non-ephys data, and all null values
                .drop(['Temp', 'Age', 'PubYear'], axis=1))
CA1_ephys_df = CA1_ephys_df.fillna(CA1_ephys_df.mean())     # Fill all null values with the column's mean
ENT_ephys_df = mouse_ephys_df[mouse_ephys_df['NeuronType'] == 'Medial entorhinal cortex layer III pyramidal cell']     # Only keeping data relevant to Medial Entorhinal Cortex
ENT_ephys_df = (ENT_ephys_df.dropna(how='all', axis=1).select_dtypes(exclude=['object'])     # Drop columns with strings, non-ephys data, and all null values
                .drop(['Temp', 'Age', 'PubYear'], axis=1))
ENT_ephys_df = ENT_ephys_df.fillna(ENT_ephys_df.mean())     # Fill all null values with the column's mean
```
